```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Reading and storing the data in a dataframe
filename = 'weather.csv'
weather = pd.read_csv(filename)
weather.head()
```

```python
# Exploring the dataframe
print(weather.shape, '\n')
print(weather.columns, '\n')
print(weather.isna().sum(), '\n')
weather.dtypes
```

```python
# Changing the datatype of date column from object to datetime
date = weather['DATE']
weather['DATE'] = pd.to_datetime(date)
weather.dtypes
```

```python
# Deleting station column as all the values correspond to a single station - USW00014765
print(weather['STATION'].value_counts())
weather.drop('STATION', axis = 'columns', inplace = True)
weather.head()
```

```python
# Describing the temperature columns
tempCols = ['TMIN', 'TAVG', 'TMAX']
print(weather[tempCols].describe())

# Examining the temperature columns using boxplot
weather[tempCols].plot(kind = 'box', title = 'Boxplot for temperature columns')

# The temperature columns look reasonable as the TAVG values are between TMIN and TMAX
```

```python
# Creating a histogram to validate if the TMAX values are greater than TMIN values
fig, ax = plt.subplots()
weather['TDIFF'] = weather['TMAX'] - weather['TMIN']
print(weather['TDIFF'].describe())
ax.hist(weather['TDIFF'], bins = 20)
ax.set_xlabel('Temperature Difference')
ax.set_ylabel('Frequency')
ax.set_title('Histogram showing difference in temperature')

# The histogram for temperature difference looks more or less normal. Moreover, it has no negative values. 
# This also indicates temperature data to be resonable. Also, the difference in the temperatures are approximately
# equally distributed with a mean value close to 18.
```

```python
# Describe the wind speed columns
windSpeed = ['AWND', 'WSF2']
weather[windSpeed].describe()

# Visualizing wind speed using boxplot
fig, ax = plt.subplots()
ax.boxplot([weather['AWND'], weather['WSF2']])
ax.set_xticklabels(['AWND', 'WSF2'])
ax.set_ylabel('Frequency')
ax.set_title('Boxplot for wind speed')
```

```python
# Creating a histogram to validate if the WSF2 values are greater than AWND values
fig, ax = plt.subplots()
weather['WDIFF'] = weather['WSF2'] - weather['AWND']
print(weather['WDIFF'].describe())
ax.hist(weather['WDIFF'], bins = 30)
ax.set_xlabel('Wind Speed Difference')
ax.set_ylabel('Frequency')
ax.set_title('Histogram showing difference in wind speed')

# The histogram looks more or less normal. Moreover, it has no negative values. 
# This also indicates the wind speed data to be resonable. 
# Also, looking at the difference in the wind speeds we could say that majority
# of the entries have difference greater than the mean value (which is close to 11).
```

```python
# Visualising bad weather conditions using histogram
badWeatherConds = weather.loc[:,'WT01':'WT22']
badWeatherCols = ['WT01', 'WT02', 'WT03', 'WT04', 'WT05', 'WT06', 'WT07',
                  'WT08', 'WT09', 'WT10', 'WT11', 'WT13', 'WT14', 'WT15',
                  'WT16', 'WT17', 'WT18', 'WT19', 'WT21', 'WT22']
weather.drop(badWeatherCols, axis = 'columns', inplace = True)

# Replacing the missing values(Nan) with 0 and counting the number of bad weather conditions for each row
badWeatherConds.fillna(0, inplace = True)
badWeatherConds['count'] = badWeatherConds.sum(axis = 'columns')

# Creating a histogram
fig, ax = plt.subplots()
ax.hist(badWeatherConds['count'], bins = 20)
ax.set_xlabel('Bad Weather Conditions Count')
ax.set_ylabel('Frequency')
ax.set_title('Histogram showing count of bad weather conditions')

# It looks like many days(approx. 44%) didn't have any bad weather conditions
```

```python
# Creating an additional column for mapping weather conditions
print(badWeatherConds['count'].unique(), '\n')
badWeatherDict = {
    0 : 'Good',
    1 : 'Bad',
    2 : 'Bad',
    3 : 'Bad',
    4 : 'Bad',
    5 : 'Worse',
    6 : 'Worse',
    7 : 'Worse',
    8 : 'Worse',
    9 : 'Worse'
}

badWeatherConds['weatherCondition'] = badWeatherConds['count'].map(badWeatherDict)
badWeatherConds.head()
```

```python
# Inserting the newly created weather condition column in weather dataframe 
# and updating its datatype from object to category
weather['weatherCondition'] = badWeatherConds['weatherCondition']
print(weather['weatherCondition'].dtype, '\n')
weather['weatherCondition'] = weather['weatherCondition'].astype('category')
print('Updated Data type: ', weather['weatherCondition'].dtype, '\n')

# Now that we have finalized our dataframe, loading this dataframe to a new CSV file in order 
# to combine it with trafficStops dataframe for further analysis
weather.to_csv('updatedWeather.csv')

weather.head(10)
```

```python
# Setting Date as index
weather.set_index('DATE', inplace = True)
weather.sort_index(level = ['DATE'], ascending = True)
weather.head()
```

```python
# Extracting month and year from date adding them as new columns into the dataframe
dict_month = {
    1 : 'Jan',
    2 : 'Feb',
    3 : 'Mar',
    4 : 'Apr',
    5 : 'May',
    6 : 'Jun',
    7 : 'Jul',
    8 : 'Aug',
    9 : 'Sep',
    10 : 'Oct',
    11 : 'Nov',
    12 : 'Dec'
}
weather['month'] = weather.index.month.map(dict_month)
weather['year'] = weather.index.year
weather.head()
```

```python
# Calculating weather condition for each month in a year
month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
yearMonthWeatherCond = weather.groupby(['year', 'month', 'weatherCondition']).agg({'weatherCondition' : 'count'})
yearMonthWeatherCond.reindex(month, level = 1)
```

```python
# Calculating monthly count of good, bad and worse weather conditions
fig, ax = plt.subplots()
good = weather[weather['weatherCondition'] == 'Good']
bad = weather[weather['weatherCondition'] == 'Bad']
worse = weather[weather['weatherCondition'] == 'Worse']

goodWeather = good.groupby('month').agg({'weatherCondition': 'count'}).reindex(month)
badWeather = bad.groupby('month').agg({'weatherCondition': 'count'}).reindex(month)
worseWeather = worse.groupby('month').agg({'weatherCondition': 'count'}).reindex(month)

ax.bar(worseWeather.index, worseWeather['weatherCondition'], label = 'Worse', color = 'red')
ax.bar(badWeather.index, badWeather['weatherCondition'], label = 'Bad', color = 'orange', bottom = worseWeather['weatherCondition'])
ax.bar(goodWeather.index, goodWeather['weatherCondition'], label = 'Good', color = 'green', bottom = badWeather['weatherCondition'] + worseWeather['weatherCondition'])
ax.set_xlabel('Month')
ax.set_ylabel('Count')
ax.set_title('Graph showing monthly count of good, bad and worse weather conditions')
ax.legend()
```

```python
# Calculating monthly minimum average, maximum average and average temperature
aggDict = {
    'TMIN' : 'mean',
    'TAVG' : 'mean',
    'TMAX' : 'mean'
}
monthlyAvgTemp = weather.groupby('month').agg(aggDict).reindex(month)
print(monthlyAvgTemp, '\n')

# Line Graph to show monthly minimum average, maximum average and average temperature
fig, ax = plt.subplots()
ax.plot(monthlyAvgTemp.index, monthlyAvgTemp['TMIN'], label = 'TMIN')
ax.plot(monthlyAvgTemp.index, monthlyAvgTemp['TAVG'], label = 'TAVG')
ax.plot(monthlyAvgTemp.index, monthlyAvgTemp['TMAX'], label = 'TMAX')
ax.set_xlabel('Month')
ax.set_ylabel('Average Temperature')
ax.set_title('Graph showing monthly minimum average, average and maximum average temperature')
ax.legend()
```

```python
# Calculating daily minimum average, maximum average and average temperature
dailyAverageTemp = weather.groupby(weather.index.day).agg(aggDict)

# Line Graph to show daily minimum average, maximum average and average temperature
fig, ax = plt.subplots()
ax.plot(dailyAverageTemp.index, dailyAverageTemp['TMIN'], label = 'TMIN')
ax.plot(dailyAverageTemp.index, dailyAverageTemp['TAVG'], label = 'TAVG')
ax.plot(dailyAverageTemp.index, dailyAverageTemp['TMAX'], label = 'TMAX')
ax.set_xlabel('Day')
ax.set_ylabel('Average Temperature')
ax.set_title('Graph showing daily minimum average, average and maximum average temperature')
ax.legend()
```
